```python
import requests
from bs4 import BeautifulSoup
import numpy as np  # Import numpy for NaN

# Replace 'main_links' with your actual list of links
main_links = ['https://nuforc.org/webreports/ndx1AL.html', 'https://nuforc.org/webreports/ndx2AL.html', 'https://nuforc.org/webreports/ndx3AL.html']

report_link = []
date = []
city = []
state = []
country = []
shape = []
duration = []
description = []
image_yn = []

for link in main_links:
    response = requests.get(link)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all the <td> tags within a parent container (e.g., <tr> or other tag)
        try:
            rows = soup.find_all('tr')  # You may need to adjust the parent container tag if not <tr>
        except AttributeError:
            print(f"Error finding data for {link}. Could not find <tr> tags.")
            continue  # Move on to the next link
        
        for row in rows:
            td_tags = row.find_all('td')
            
            if len(td_tags) >= 1:  # Make sure there are at least 9 <td> tags containing the data you need
                date_val = td_tags[0].get_text()
                city_val = td_tags[1].get_text()
                state_val = td_tags[2].get_text()
                country_val = td_tags[3].get_text()
                shape_val = td_tags[4].get_text()
                duration_val = td_tags[5].get_text()
                description_val = td_tags[6].get_text()
                image_yn_val = td_tags[8].get_text()  # Adjust index based on the column you need
                
                # Check for empty <td> tags and replace with NaN
                if not date_val.strip():
                    date_val = np.nan
                if not city_val.strip():
                    city_val = np.nan
                if not state_val.strip():
                    state_val = np.nan
                # ... Repeat the above for other columns
                
                date.append(date_val)
                city.append(city_val)
                state.append(state_val)
                country.append(country_val)
                shape.append(shape_val)
                duration.append(duration_val)
                description.append(description_val)
                image_yn.append(image_yn_val)
            else:
                print(f"Data format on {link} is incorrect.")
                print(f"Number of <td> tags found: {len(td_tags)}")
                print(td_tags)  # Print the <td> tags to inspect the content
    else:
        print(f"Error finding data for {link}. Status Code: {response.status_code}")

# Now you have all the lists filled with the extracted data. You can proceed to create a DataFrame if needed.
```

```python
table_url = 'https://nuforc.org/webreports/ndxlAL.html'
table_response = requests.get(table_url)
table_soup = BS(table_response.text)
table_soup
```

```python
table_soup.findAll('tr')('td')
```

```python

rows = table_soup.find_all('tr')

data = []

for row in rows:
    td_tags = row.find_all('td')
    
    row_data = []
    
    for td in td_tags:
        a_tag = td.find('a')
        if a_tag is not None:
            link = a_tag['href']
            link_text = a_tag.get_text()
            row_data.append(link_text)
            row_data.append(link)

        row_data.append(td.get_text())

    data.append(row_data)


```
